# Aprendizagem não supersionado

##  Diferenças: 

### Supervisionado
Treinamento com entradas e saidas definidas -> 
- MLP
- XGBoost
- LSTM
- CNN
- RNN

### Não Supervisionado
Treinamento com entradas e saidas não definidas -> 
- Agrupamento (K-means, DBSCAN, Hierarchial) 
- Regras de Associação (Eclat, Apriori)
- Densidade de Probabilidade ()
- Redução de Dimensionalidade (T-SNE)

```python
# Clusterização

# Agrupa dados semelhantes
# Não precisa de rotulos (labels)
# Descobre padrões ocultos

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#Dados de entrada
X = np.array([
    [1, 2], [1, 3], [2, 2],
    [8, 8], [9, 8], [8, 9],  
])
```

```python
# Aplicar o KMeans
                #Qtd de Grupos
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
centroids = kmeans.cluster_centers_

#Visualização
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centroids[:,0], centroids[:,1], marker='+', s=200)
plt.show()
```
