```python
#Exemplo Funcionamento de uma rede neural MLP

import numpy as np

#camada de entrada
x = np.array([1,2])

#camada oculta
w1 = np.array([
    [0.5, -0.4],
    [0.3, -0.8],
    [0.6, -0.2]
])

b1 = np.array([ 0.1, -0.2, 0.05 ])

#Funcao de ativação relu
def relu(z):
    return np.maximum(0, z)

#calculo da camada oculta
z1 = w1 @ x + b1
a1 = relu(z1)

#camada de saida
w2 = np.array([0.7, -0.5, 0.9])
b2 = 0.2

#calculo da saida
z2 = w2 @ a1 + b2


#fução de ativação da camada de saida
def sigmoide(z):
    return 1/ (1+ np.exp(-z))

#Saida final
yregressao = z2
yclassificacao = sigmoide(z2)
print(f"Saída para regressão = {yregressao}")
print(f"Saída para classificação = {yclassificacao}")
```
