```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
# Carregar e preparar os dados
iris = load_iris()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)
```

```python
# Parêmtros do modelos, labels e score
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)
score = silhouette_score(X_scaled, labels)
score
```

```python
# Clusters encontrados
plt.figure(figsize=(8, 6))
plt.scatter(iris.data[:, 0], # Comprimento da sépala
iris.data[:, 1], # Largura da sépala
c=labels, # Cor pelo cluster
cmap='viridis')
plt.xlabel('Comprimento da Sépala (cm)', fontsize=12)
plt.ylabel('Largura da Sépala (cm)', fontsize=12)
plt.title(f'Clusters Encontrados pelo K-Means (K=3)\nSilhueta = {score:.3f}')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Classes reais
plt.figure(figsize=(6, 4))
plt.scatter(iris.data[:, 0], # Comprimento da sépala
iris.data[:, 1], # Largura da sépala
c=iris.target, # Cor pela classe real
cmap='viridis')
plt.xlabel('Comprimento da Sépala (cm)', fontsize=12)
plt.ylabel('Largura da Sépala (cm)', fontsize=12)
plt.title('Classes Reais das Flores (Setosa,
Versicolor, Virginica)')
# Criar a barra de cores com os nomes das espécies
cbar = plt.colorbar(label='Espécie')
cbar.set_ticks([0, 1, 2])
cbar.set_ticklabels(['Setosa', 'Versicolor',
'Virginica'])
plt.show()
```
