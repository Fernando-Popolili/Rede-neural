```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Dataset de fraude via OpenML
data = fetch_openml(name="creditcard", version=1, as_frame=True)

df = data.frame.copy()

# A coluna alvo costuma ser "Class"
X = df.drop("Class", axis=1)
y = df["Class"].astype(int)

print("\nDistribuição das classes:")
print(y.value_counts())

# PADRONIZAR OS DADOS
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# APLICAÇÃO DO PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Transformar em DataFrame para facilitar
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_pca["Class"] = y.values

# PLOT
plt.figure(figsize=(6, 4))

# Classe 0 = normal
plt.scatter(
    df_pca[df_pca["Class"] == 0]["PC1"],
    df_pca[df_pca["Class"] == 0]["PC2"],
    alpha=0.3,
    label="Normal",
    s=10)

# Classe 1 = fraude
plt.scatter(
    df_pca[df_pca["Class"] == 1]["PC1"],
    df_pca[df_pca["Class"] == 1]["PC2"],
    alpha=0.8,
    label="Fraude",
    s=20)

plt.title("PCA aplicado ao dataset de fraude")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend()
plt.grid(True)
plt.show()

import numpy as np
from sklearn.decomposition import PCA

# PCA sem limitar componentes
pca_full = PCA()
pca_full.fit(X_scaled)

# Variância acumulada
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)

# Encontrar número de componentes para 95%
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1

print("Número de componentes para 95% da variância:", n_components_95)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(cumulative_variance, marker='o')
plt.axhline(y=0.95, linestyle='--', label='95% da variância')

plt.xlabel("Número de Componentes")
plt.ylabel("Variância Acumulada")
plt.title("Quantas componentes explicam 95% da variância")
plt.legend()
plt.grid(True)

plt.show()
```
